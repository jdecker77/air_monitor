```python
"""
Merge census data with shape file
"""

def MergeFrames(shapes,domain,join_ID):
    
    # Merge data into geo file
    gdf = shapes.merge(domain,on=join_ID,sort=True)
    gdf.crs = shapes.crs
    
    return gdf
```

```python
"""
Merge census data with shape file

Census includes ['ID','Variable']
"""

def MergeCensusFrames(shapes,domain,join_ID):
    
    new_cols = domain['Variable'].tolist()
    domain = domain.drop(labels = ['ID','Variable'],axis=1)

    domain = domain.T

    old_cols = domain.columns.tolist()
    columns = dict(zip(old_cols,new_cols))
    domain = domain.rename(columns=columns)

    names = []
    for i in domain.index.tolist():
        if i[-1] == " ":
            names.append(i[:-1])
        else:
            names.append(i)
    domain['NAME'] = names
    
    # Merge data into geo file
    gdf = shapes.merge(domain,on=join_ID,sort=True)
    gdf.crs = shapes.crs
    
    return gdf
```

```python
# Select rows by ID-1
census_age_combined = census_age.iloc[1:25]
census_age_combined
```

```python
census_age_male = census_age.iloc[27:51]
census_age_male
```

```python
census_age_female = census_age.iloc[53:78]
census_age_female
```

```python
total_pop = dict(census_age_combined.sum(axis=0)[2:])
male_pop = dict(census_age_male.sum(axis=0)[2:])
female_pop = dict(census_age_female.sum(axis=0)[2:])

columns = ['total_pop','male_pop','female_pop']

index = list(np.arange(0,len(total_pop.values())))

df = pd.DataFrame(data = [total_pop,male_pop,female_pop]).T

df = df.rename(columns={0: 'total_pop',1:'male_pop',2:'female_pop'})

names = []
for i in list(total_pop.keys()):
    if i[-1] == " ":
        names.append(i[:-1])
    else:
        names.append(i)
df['NAME'] = names

# Add categorical column
genders = []
for row in range(0,len(names)):
    gender = ''
    if df.iloc[row]['male_pop'] > df.iloc[row]['female_pop']:
        genders.append('Male')
    else:
        genders.append('Female')

GENDER_MAJORITY = pd.Series(genders)
df['Gender_Majority'] = GENDER_MAJORITY.values

gdf = MergeFrames(local_area_bounds,df,'NAME')

gdf['area'] = gdf.area


gdf.to_file('../data/Vancouver/census/output/van_city_pop_totals/van_city_pop_totals.shp',index=True)
gdf.head()
```
