```python
'''
Save graph as both shapefile and graph
'''

def SaveGraph(H,filename):
    ox.save_graph_shapefile(H,filename=filename,folder=GetNetFolder())
    ox.save_graphml(H,filename=filename+'.graphml',folder=GetNetFolder())
```

```python
'''
Convert edge/nodes dataframes to graph
'''

def FramesToGraph(edges,nodes):
    return ox.gdfs_to_graph(nodes, edges)
```

```python
'''
Convert graph to edge/nodes dataframes
'''

def GraphToFrames(H):
    return ox.graph_to_gdfs(H)
```

```python
'''
Load graph from file
'''
def LoadSavedGraph(filename):
    return ox.load_graphml(filename='network_5km.graphml',folder=GetNetFolder())
```

```python
'''
Get new graph from point

Use network_type or custom filter, custom_filter='["highway"~"cycleway"]'
'''

def GetNewGraph(net_type):
    
    point = (float(station.iloc[0]['LATITUDE']),float(station.iloc[0]['LONGITUDE']))

    gph = ox.graph_from_point(point,5000,distance_type='network',network_type=net_type,simplify=True,infrastructure='way["highway"]')
    gph = ox.remove_isolated_nodes(gph)
    
    return gph
```

```python
'''
Open traffic counts
'''

month = 10
day = 11
hour = 11

filename = GetImageRoot(month,day,hour)+'/'+'traffic_measurement.csv'
# print(GetImageRoot(month,day,hour))

images = pd.read_csv(filename)
images = images.drop('Unnamed: 10',axis=1)

images['camid'] = images['ImagePath'].map(lambda x: x[-9:-6])
images['camid'] = images['camid'].astype(int)

images = images.sort_values('camid')

images.head()
```

```python
'''
merge osmid into traffic counts
'''

merged = images.merge(camlocs,on='camid')
merged.head()
```

```python
'''
Get count of vehicle types by node
'''

counts = merged.groupby('osmid').sum()

cols = ['bicycle', 'car', 'motorcycle',
       'bus', 'train', 'truck']
counts = counts[cols]

counts.reset_index(drop=False,inplace=True)

counts
```

```python
'''
Determine emission load at interesction
'''

cols = ['bicycle', 'car', 'motorcycle',
       'bus', 'train', 'truck']

values = [-1,1,1,3,5,2]

combos = dict(zip(cols,values))

rows = counts.index.tolist() 
for row in rows:
    for col in cols:
        counts.at[row,col] = counts.at[row,col]*combos[col]
```
