```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import geopandas as gpd
from shapely.geometry import Point, LineString,Polygon

import seaborn as sns
```

```python
'''
Dfine polygon for comparison
'''

def DefinePoly():
    # Vic     
    ll = (-123.39693,48.40227)
    tl = (-123.39595,48.45085)
    tr = (-123.31474,48.45224)
    lr = (-123.31340,48.40162)
    
    return Polygon([ll,tl,tr,lr])
```

```python
'''
Convert points to gdf - GIS
'''

def CoordToPoint(data):
    
    #zip the coordinates into a point object and convert to a GeoData Frame
    geometry = [Point(xy) for xy in zip(data.LONGITUDE, data.LATITUDE)]
    geo_df = gpd.GeoDataFrame(data, geometry=geometry)
    geo_df.crs = {'init' :'epsg:4326'}

    return geo_df
```

```python
stations = pd.read_csv('../data/ems/bc_air_monitoring_stations.csv',header=0,encoding = "ISO-8859-1")

stations['EMS_ID'] = stations['EMS_ID'].astype(str)
stations = stations[stations['STATUS']==1]

cols = ['EMS_ID','LATITUDE','LONGITUDE']
stations = stations[cols]

stations.head()
```

```python
l = stations['EMS_ID'].tolist()
newstats = [] 
for i in l:
    if i[0].isnumeric():
        i = 'U'+i
#         print(i)
    newstats.append(i)
    
stations['EMS_ID'] = newstats
```

```python
'''
Open source file
'''

air = pd.read_csv('../data/ems/NO2.csv',header=0,encoding = "ISO-8859-1")

air['EMS_ID'] = air['EMS_ID'].astype(str)

air.head()
```

```python
data = air.merge(stations, on='EMS_ID')
data.head()
```

```python
'''
Convert to gdf
'''

gdf = CoordToPoint(data)
```

```python
'''
Limit data by area
'''

table = gdf[gdf['geometry'].map(lambda x:x.within(DefinePoly()))]

table.head()
```
