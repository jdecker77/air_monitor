```python
'''
Split, fit, and evaluate models
'''

def FitAndScoreREG(features,labels,regressors):
    
    for regressor in regressors:
        print('-----------------------------------------------------------------')
        print(regressor['regressor_name'])
        
        # Fit each model         
        reg = regressor['regressor_func']
        
        rmse = cross_val_score(reg, features, labels.ravel(), scoring='neg_mean_squared_error', cv=10)
        print('Cross Validation RMSE',rmse.mean())
        print('\n')
        
        # Fit model to check weights
        reg.fit(features,labels)
        
        # View theta values         
        print('Weights:',list(reg.coef_))
        print('Mean Weights:',reg.coef_[0].mean())
        print('Intercept:',reg.intercept_)
        
        print('\n')
```

```python
'''
Check for colinearity/multicolinearity

Variance Inflation Factor (VIF) is a measure of colinearity among predictor variables within a multiple regression. 
VIF estimates how much the variance of a regression coefficient is inflated due to multicollinearity in the model.

V.I.F. = 1 / (1 - R^2)

1 = not correlated.
Between 1 and 5 = moderately correlated.
Greater than 5 = highly correlated.


Because several features show VIF > 5, multicolinearity may impact the model. I will use regularized algorithms to build the models.
'''

features = "+".join(df.columns[:-1])
# print(features)

# get y and X dataframes based on this regression:
target, feats = dmatrices('NO2 ~' + features, df, return_type='dataframe')

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(feats.values, i) for i in range(feats.shape[1])]
vif["features"] = feats.columns

vif.sort_values(by='VIF Factor',ascending=False).round(1)
```

### Transform Data

```python
'''
Normalize float values to 0/1 for columns numeric coumns
'''

scaler = MinMaxScaler()

# Fit and transform values back into df
cols = df.columns.tolist()[:-1]
df[cols] = scaler.fit_transform(df[cols].values)

df.head()
```

### Create Dataset

```python
'''
Get features/labels as np arrays
'''

X = df.values[:,:-1]
y = df.values[:,-1:]
```

### Build Model

```python
'''
Define regression models
'''

OLS = {'regressor_name':'OLS','regressor_func': linear_model.LinearRegression()}
RDG = {'regressor_name':'RDG','regressor_func': linear_model.Ridge(alpha=1.0)}
LAS = {'regressor_name':'LAS','regressor_func': linear_model.Lasso(alpha=1.0)}
ELA = {'regressor_name':'ELA','regressor_func': ElasticNet(alpha=0.5,random_state=0)}
SGD = {'regressor_name':'SGD','regressor_func': linear_model.SGDRegressor(max_iter=1000, tol=1e-3)}
```

```python
'''
Fit and evaluate model - All Features
'''

regressors = [OLS,RDG,LAS,ELA,SGD]

FitAndScoreREG(X,y,regressors)
```
