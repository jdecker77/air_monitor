```python
def StartTimer(month,day,hour,minutes=0):
    start = datetime(2019,month,day,hour,minutes)
#     print(start)
#     print(datetime.now())
    if start > datetime.now():
        delta = int((start - datetime.now()).total_seconds())
        print('Waiting',delta,'seconds.')
        sleeper.sleep(delta)
        return start
    else:
        print('Error in start time.')
```

```python
'''
Add month and day folders if needed
'''

def WriteFolder(month,day,hour,cameraID):
    APP_ROOT = os.path.dirname(os.path.abspath('Camera_collection.ipynb'))

    cam_folder = GetCamFolder().replace('../','')

    MONTH_ROOT = APP_ROOT.replace('notebooks',cam_folder+'/M'+str(month))
    # print(MONTH_ROOT)
    if not os.path.isdir(MONTH_ROOT):
        os.mkdir(MONTH_ROOT)

    DAY_ROOT = MONTH_ROOT+"/D"+str(day)
    # print(DAY_ROOT)
    if not os.path.isdir(DAY_ROOT):
        os.mkdir(DAY_ROOT)

    HOUR_ROOT = DAY_ROOT+"/H"+str(hour)
    # print(HOUR_ROOT)
    if not os.path.isdir(HOUR_ROOT):
        os.mkdir(HOUR_ROOT)

    # Add cameraID folder
    CAM_ROOT = HOUR_ROOT+"/C"+str(cameraID)
    # print(CAM_ROOT)
    if not os.path.isdir(CAM_ROOT):
        os.mkdir(CAM_ROOT)
```

```python
'''
Request and save a single image
'''

def GetPhoto(url,filename):
    
    try:
        with closing(request.urlopen(url)) as r:
            with open(filename, 'wb') as f:
                shutil.copyfileobj(r, f)
    except Exception as e:
        print(e)
```

```python
def IncrementTime(month,day,hour):
    
    try:
        newTime = str(datetime(2019,month,day,hour)+timedelta(hours=1))
    except ValueError as e:
        if str(e) == "hour must be in 0..23":
            hour = 0
            day += 1
            try:
                newTime = datetime(2019,month,day,hour)
            except ValueError as e:
                if str(e) == "day is out of range for month":
                    month += 1
                    day = 1
                try:
                    datetime(2019,month,day,hour)
                except ValueError as e:
                    if str(e) == "month must be in 1..12":
                        month = 1
                        newTime = datetime(2019,month,day,hour)
    
    month = int(newTime[5:7])
    day = int(newTime[8:10])
    hour = int(newTime[11:13])
                
    return month,day,hour
```

```python
def Sleep(secs):
    print('Sleeping',secs,'seconds')
    sleeper.sleep(secs)
```

```python
'''
Primary Loop 

loop through list of camera ids to get new images

Add in timer
add in schedule month/day/hour
'''

cams = [200,498,499,507,508,509,510,511,512,513,514,855,856,857,858]
# cams = [200]

# Set  start params
month = 10
day = 20
hour = 17

calls_per_hour = 4

# Start a timer here. sleep until above params
start = StartTimer(month,day,hour,minutes=2)
print('Starting Collection at', datetime.now())

stop = start+timedelta(days=1)
while datetime.now() < stop:
    
    # Loop through calls per hour
    calls = list(np.arange(1,calls_per_hour+1))
    for call in calls[:]:
        for camID in cams:
            
            # Set camera URL 
            if camID == 200:
                url = 'http://roofcam.royalbcmuseum.bc.ca:8888/axis-cgi/mjpg/video.cgi'
            else:
                url = 'https://images.drivebc.ca/bchighwaycam/pub/cameras/'+str(camID)+'.jpg'
        #     print(url)

            # Write folder if needed
            WriteFolder(month,day,hour,camID)

            # Set filename
            camFolder = GetImageFolderName(month,day,hour,camID)
            filename = camFolder+'/'+str(camID)+'_'+str(call)+'.jpg'
#             print(filename)
    
            # Write photo
            if camID == 200:
                GetVideo(url,filename)
            else:
                GetPhoto(url,filename)
        
        # Sleep for interval in seconds         
        Sleep(3600/calls_per_hour)
        
    # Increment by hour         
    month,day,hour = IncrementTime(month,day,hour)
    
print('Stopping Collection')
```

```python
'''
Save frame 
'''

def GetVideo(url,filename):
    
    r = requests.get(url, stream=True)    
    if(r.status_code == 200):
        tmp = bytes()
        for chunk in r.iter_content(chunk_size=1024):
            tmp += chunk
            a = tmp.find(b'\xff\xd8')
            b = tmp.find(b'\xff\xd9')
            if a != -1 and b != -1:
                jpg = tmp[a:b+2]
                tmp = tmp[b+2:]
                frame = cv2.imdecode(np.fromstring(jpg, dtype=np.uint8), cv2.IMREAD_COLOR)
                
                # Crop frame
                y=325
                x=500
                h=315
                w=480
                crop_img = frame[y:y+h, x:x+w]
                cv2.imwrite(filename,crop_img)
                break
            else:
                pass
#                 print('No jpg')
    else:
        print("Received unexpected status code {}".format(r.status_code))
```
